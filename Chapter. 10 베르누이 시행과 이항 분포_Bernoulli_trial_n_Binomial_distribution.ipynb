# 베르누이 시행과 이항분포

```python
import numpy as np
import matplotlib.pyplot as plt

def binomial_pmf(n, p, k):
    """이항 분포의 확률질량함수(PMF) 계산"""
    return np.math.comb(n, k) * p**k * (1-p)**(n-k)

def binomial_cdf(n, p, k):
    """이항 분포의 누적 분포 함수(CDF) 계산"""
    cdf = 0
    for i in range(k + 1):
        cdf += binomial_pmf(n, p, i)
    return cdf

# Simulate data
np.random.seed(123)

# Define parameters
n = 100
p = 0.3
Y = np.random.binomial(1, p, n)
# Calculate PMF and CDF
k_values = np.arange(0, n+1)
pmf = [binomial_pmf(n, p, k) for k in k_values]
cdf = [binomial_cdf(n, p, k) for k in k_values]

# Plot the data
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Left subplot: PMF and CDF of binomial distribution
color = 'tab:blue'
ax1.plot(k_values, pmf, color=color, marker='o', label='PMF')
ax1.set_xlabel('k')
ax1.set_ylabel('PMF', color=color)
ax1.tick_params(axis='y', labelcolor=color)
ax1.grid(True, linestyle='--')

ax1_twin = ax1.twinx()
color = 'tab:red'
ax1_twin.plot(k_values, cdf, color=color, marker='x', linestyle='--', label='CDF')
ax1_twin.set_ylabel('CDF', color=color)
ax1_twin.tick_params(axis='y', labelcolor=color)

ax1.set_title(f'Binomial Distribution (n={n}, p={p})')
ax1.grid(True, linestyle='--')
ax1.legend(loc='upper left')
ax1_twin.legend(loc='upper right')

# Right subplot: Histogram of data Y
ax2.hist(-Y, bins=2, color='b', edgecolor='white', alpha=0.6)
ax2.set_xlabel('Outcome')
ax2.set_ylabel('Frequency')
ax2.set_xticks([-1, 0])
ax2.set_xticklabels(['Success', 'Failure'])
ax2.set_title('Histogram of Data Y')
ax2.grid(True, linestyle='--')

plt.tight_layout()
plt.show()
```
